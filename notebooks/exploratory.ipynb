```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df = pd.read_csv('../data/weatherHistory.csv')

# Show first 5 rows
df.head()
```

```python
# Get summary info
df.info()

# Describe numerical data
df.describe()
```

```python
# Check missing values
df.isnull().sum()
```

```python
plt.figure(figsize=(10, 4))
plt.plot(pd.to_datetime(df['Formatted Date']), df['Temperature (C)'])
plt.title('Temperature Over Time')
plt.xlabel('Date')
plt.ylabel('Temperature (C)')
plt.grid(True)
plt.show()
```

```python
df['Formatted Date'] = pd.to_datetime(df['Formatted Date']).dt.tz_localize(None)

plt.figure(figsize=(10, 4))
plt.plot(df['Formatted Date'], df['Temperature (C)'])
plt.title('Temperature Over Time')
plt.xlabel('Date')
plt.ylabel('Temperature (C)')
plt.grid(True)
plt.show()
```
